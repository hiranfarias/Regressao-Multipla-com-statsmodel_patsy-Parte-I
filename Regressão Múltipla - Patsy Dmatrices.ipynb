```python
import statsmodels.formula.api as smf
from seaborn import load_dataset
import matplotlib.pyplot as plt
import statsmodels.api as sm
from patsy import dmatrix
import seaborn as sns
import pandas as pd
import numpy as np
import patsy
```

```python

df = pd.read_csv('previsao_de_renda.csv')

df = pd.read_csv('previsao_de_renda.csv').drop(['Unnamed: 0', 'id_cliente'], 
                                               axis = 1).dropna().drop_duplicates().reset_index(drop = True)
```

1. Ajuste um modelo para prever log(renda) considerando todas as covariáveis disponíveis.
    - Utilizando os recursos do Patsy, coloque as variáveis qualitativas como *dummies*.
    - Mantenha sempre a categoria mais frequente como casela de referência
    - Avalie os parâmetros e veja se parecem fazer sentido prático.  


```python

y, x = patsy.dmatrices('''
                            renda ~ 
                            sexo + 
                            posse_de_imovel + 
                            idade + 
                            tempo_emprego + 
                            C(estado_civil) + 
                            C(tipo_residencia) + 
                            qtd_filhos + 
                            qt_pessoas_residencia + 
                            C(educacao) + 
                            posse_de_veiculo + 
                            data_ref
                        ''' , 
                       data = df)

sm.OLS(y, x).fit().summary()

#dmatrix('C(posse_de_imovel)', df)
```

2. Remova a variável menos significante e analise:

```python

y, x = patsy.dmatrices('''
                            renda ~ 
                            sexo + 
                            posse_de_imovel + 
                            idade + 
                            tempo_emprego + 
                            C(estado_civil) + 
                            C(tipo_residencia) + 
                            qtd_filhos + 
                            qt_pessoas_residencia + 
                            C(educacao) + 
                            posse_de_veiculo
                            
                        ''', data = df)

sm.OLS(y, x).fit().summary()

# Adj. R-squared  --> aumentou
# pvalue de posse_de_imovel --> baixou
# pvalue C(tipo_residencia)[T.Casa] --> baixou
# pvalue C(tipo_residencia)[T.Com os pais] --> baixou
# pvalue C(tipo_residencia)[T.Comunitaria] --> baixou
# # pvalue C(tipo_residencia)[T.Estúdio] --> aumentou
# pvalue C(tipo_residencia)[T.Governamental] --> aumentou
# C(educacao) --> baixou de forma geral
# pvalue posse_de_veiculo[T.True] --> baixou
# idade e tempo_emprego se mantiveram com os mesmos valores
# pvalue qtd_filhos --> baixou
# pvalue qt_pessoas_residencia --> baixou

# uma melhora bem sútiul, mais melhorou
```
